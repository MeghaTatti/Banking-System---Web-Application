# Get BestBuy timeline for the screen-name called "BestBuy_Deals".

```python
# Get BestBuy Deals timeline and screen name.
screen_name = 'ChicagoBankRate'
timeline = [tweet for tweet in twitter.request('statuses/user_timeline',
                                                {'screen_name': screen_name,
                                                 'count': 200})]
print ('got %d tweets for user %s' % (len(timeline), screen_name))
```

```python
# Print time got created.
timeline[3]['created_at']
```

```python
# Print the text.
print ('\n\n\n'.join(t['text'] for t in timeline))
```

```python
import re
for tweet in timeline:
    news = tweet['text']
    print( news + '\n')
```

```python
bankNews=[]

for tweet in timeline:
    news = (tweet['text']).encode('ascii','ignore').decode()
    bankNews = bankNews + [news]
```

```python
# Sanity Test that  we got some deals
bankNews[:5]
```

# Create and write the deals into DealMatches.txt file that will be used by web-app called SmartPortables to display two deal matches.

```python
BankRate = open('BankRate.txt', 'w')

for news in bankNews[:5]:
  BankRate.write("%s\n" % news)
  print(news)

BankRate.close()
```
